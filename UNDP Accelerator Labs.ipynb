#### Create Location Dataframe

```python
# Cleanup

df_location = df[['Project ID', 'UNDP Accelerator Lab', 'Latitude', 'Longitude']]
```

```python
# Replacing 'xxxxx' to '0000' - (' ' replaced xxxxx after the first attempt)

df_location['Latitude'].replace('xxxxx','0000', inplace=True)
```

```python
# Replacing 'xxxx' to '0000' - (' ' replaced xxxxx after the first attempt)

df_location['Latitude'].replace('xxxx','0000', inplace=True)
```

```python
# Converting datatype to float from object

df_location['Latitude'] = df_location.Latitude.astype(float)
```

```python
#Replacing 'xxxxx' to '0000'

df_location['Longitude'].replace('xxxxx','0000', inplace=True)
```

```python
#Converting datatype to float from object

df_location['Longitude'] = df_location.Longitude.astype(float)
```

```python
# Calling the nominatim tool

geoLoc = Nominatim(user_agent="GetLoc")
```

```python
# Passing the coordinates

df_location['Address'] = df_location.apply(lambda row: geoLoc.reverse(f"{row['Latitude']},{row['Longitude']}").raw['address']
                                                                        ,axis=1)
```

```python
# Join the values of the Address Dictionary as string excluding country code

df_location['Address 1'] = df_location.apply(lambda row: ', '.join(list(row['Address'].values())[:-1]),axis=1)
```

```python
# Translate the address with Deep-Translator Library

df_location['Address 2'] = df_location.apply(lambda row: GoogleTranslator(source='auto', 
                            target='en').translate(row['Address 1']),axis=1)
```

```python
# Add country column based on 'UNDP Accelerator Lab'

def country(row):
    if 'AccLab' in row['UNDP Accelerator Lab']:
        return ' '.join(row['UNDP Accelerator Lab'].split()[1:])
    return 'India'
```

```python
df_location['Country'] = df_location.apply(country,axis=1)
```

```python
# Drop Address 1 column

df_location.drop('Address 1', axis=1, inplace=True)
```

```python
# Rename Address and Address 2 columns

existing_cols = {
    'Address':'Address Dictionary',
    'Address 2':'Address'
}
```

```python
df_location.rename(columns=existing_cols, inplace=True)
```

#### Create Solutions Info Dataframe

```python
# Dataframe Containing Main Solutions Info

df_UNDP_Project_Data = df[['Project ID','Date','Title','Contributor',
              'Energy Source','Clean Cooking','Solution Description','Solution URL','Open Source','IP Protected',
              'User Trainable','Prototype','Product','Available in Market',
              'Advance Order Placement','Technological Readiness Level','Diffusion Status and End-Users Feedback',
              'End-Users Feedback URL','Efficiency Outcome','Potential Challenges']]
```

```python
# Update Date in df_UNDP_Project_Data
# Update new_date to MM/DD/YYYY

df_UNDP_Project_Data['Date'] = df_UNDP_Project_Data['Date'].str[4:]
```

```python
df_UNDP_Project_Data['Date'] = df_UNDP_Project_Data['Date'].str[:-9]
```

```python
# Create pandas series for the main dataframe's Date

date_series = pd.Series(df_UNDP_Project_Data.Date)
```

```python
# The str extract method is used to extract out the month, day and year into individual pandas series

month = date_series.str.extract(r'([a-zA-Z]+)', expand=True)
```

```python
day = date_series.str.extract(r'(\d{2})', expand=True)
```

```python
year = date_series.str.extract(r'(\d{4})', expand=True)
```

```python
# Convert written month to digits

month_dict = {
            'Jan':'01', 
             'Feb':'02',
             'Mar':'03',
             'Apr':'04',
             'May':'05',
             'Jun':'06',
             'Jul':'07',
             'Aug':'08',
             'Sep':'09',
             'Oct':'10',
             'Nov':'11',
             'Dec':'12'
            }
```

```python
for key, value in month_dict.items():
    month = month.replace(key, value)
```

```python
# Build the new date field

df_UNDP_Project_Data['New Date'] = month + '/' + day + '/' + year
```

```python
# Drop initial date fields

df_UNDP_Project_Data = df_UNDP_Project_Data.drop(['Date'], axis=1)
```

```python
# Rename date column

df_UNDP_Project_Data = df_UNDP_Project_Data.rename({'New Date':'Date'}, axis=1)
```

```python
# Reorder Columns

new_columns = ['Project ID','Date','Title','Contributor',
              'Energy Source','Clean Cooking','Solution Description','Solution URL','Open Source','IP Protected',
              'User Trainable','Prototype','Product','Available in Market',
              'Advance Order Placement','Technological Readiness Level','Diffusion Status and End-Users Feedback',
              'End-Users Feedback URL','Efficiency Outcome','Potential Challenges']
```

```python
df_UNDP_Project_Data = df_UNDP_Project_Data[new_columns]
```

```python
# Update each column that has a boolean response of x with YES, and NaN with NO

bool_clean = ['Clean Cooking','Open Source','IP Protected','User Trainable',
              'Prototype','Product','Available in Market','Advance Order Placement']
```

```python
# Replacing above items with either False or True doesn't change the column's datatype to bool

for item in bool_clean:
    df_UNDP_Project_Data[item] = df_UNDP_Project_Data[item].fillna("NO").replace(['x'],'YES')
```

```python
# Fill each row that is empty for columns with free text to empty string from NaN

free_text_clean = ['Date','Energy Source','Solution Description','Solution URL',
                   'Technological Readiness Level','Diffusion Status and End-Users Feedback','End-Users Feedback URL','Efficiency Outcome',
                   'Potential Challenges']
```

```python
for item in free_text_clean:
    df_UNDP_Project_Data[item] = df_UNDP_Project_Data[item].fillna("")
```
